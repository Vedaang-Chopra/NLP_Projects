```python
#EDA on text data i.e., on the description column : data['Description']
#step-1: Converting into lower case :
import string
description=[doc.lower() for doc in data['Description']]
print(description)
```

```python
vocab=len(description)
print("Size of the vocabulary is ",vocab)
```

```python
#step-2: Tokenization & Removing punctuations :
import nltk
tokenizer = nltk.RegexpTokenizer(r"\w+")
new_words = tokenizer.tokenize(str(description))
print(new_words)
```

```python
#step-3: Removing stop words:
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = set(stopwords.words('english'))
words = [w for w in new_words if not w in stop] 
print(words)
```

Stopwords like while,the,of,for,to etc are removed from the raw input data.

```python
print(len(words))
```

```python
# Step-4: Stemming and Lemmatization
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet  import WordNetLemmatizer
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in words]
print(stemmed)
```

```python
nltk.download('wordnet')
wordnet=WordNetLemmatizer()
lem=[wordnet.lemmatize(word) for word in stemmed]
print(lem) 
```

```python
print(lem)
print(len(lem))
```

```python
from nltk.util import ngrams
unigrams=ngrams(lem,1)
for u in unigrams:
  print(u)
```

```python
from nltk.util import ngrams
bigrams=ngrams(lem,2)
for b in bigrams:
  print(b)
```
